```python
#Regression analysis between auto weight and horse power. 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

auto = pd.read_csv('C:/Users/yiche/Desktop/Python/car.csv')

print(auto.head())
```

```python
sns.lmplot(x='weight', y='hp', data=auto)

plt.show()
```

```python
sns.residplot(x='weight', y='hp', data=auto, color='green')

plt.show()
```

```python
plt.scatter(auto['weight'], auto['hp'], label='data', color='red', marker='o')

sns.regplot(x='weight', y='hp', data=auto, order=1, color='blue', scatter=None, label='First Order')

sns.regplot(x='weight', y='hp', data=auto, order=2, color='green', scatter=None, label='Second Order')

plt.legend(loc='upper right')
plt.show()
```

```python
sns.lmplot(x='weight', y='hp', data=auto, hue='origin', palette='Set1')

sns.lmplot(x='weight', y='hp', data=auto, row='origin')

plt.show()
```

```python
sns.stripplot(x='cyl', y='hp', data=auto, jitter=True, size=3)

plt.show()
```

```python
sns.swarmplot(x='hp', y='cyl', data=auto, hue='origin', orient='h')

plt.show()
```

```python
sns.violinplot(x='cyl', y='hp', data=auto, inner=None, color='lightgray')

# Overlay a strip plot on the violin plot
sns.stripplot(x='cyl', y='hp', data=auto, jitter=True, size=1.5)

plt.show()
```

```python
sns.jointplot('weight', 'hp', auto, kind='reg')
plt.show()
```
